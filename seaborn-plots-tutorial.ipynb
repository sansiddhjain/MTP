```python
import seaborn as sns
import pandas as pd
```

```python
sns.set(style="darkgrid")
```

```python
fmri = sns.load_dataset("fmri")
bf_adv = pd.read_csv("bf_adv-seaborn.csv")
del bf_adv["Unnamed: 0"]
bf_adv1 = bf_adv[bf_adv['cluster'] < 3]
```

```python
palette = sns.color_palette("hls", 3)
sns.lineplot(x="percentile", y="value", hue="cluster", palette=palette, data=bf_adv1)
# sns.lineplot(x="timepoint", y="signal", hue="region", style="event", data=fmri)
```
